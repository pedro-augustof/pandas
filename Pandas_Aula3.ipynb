```python
#Criando dataframe
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.head()
```

```python
#Quantidade de linhas e colunas
df.shape
```

```python
#Tipos de dados
df.dtypes
```

```python
#Qual a Receita total?
df["Valor Venda"].sum()
```

```python
#Criando a coluna de custo
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head(1)
```

```python
#Custo total
round(df["custo"].sum(), 2)
```

```python
#Coluna de lucro, Receita - Custo
df["lucro"] = df["Valor Venda"] - df["custo"]
```

```python
#Coluna com total de dias para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```



> Media do tempo de envio para cada Marca



```python
#Extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
#Verificando o tipo da coluna Tempo_envio
df["Tempo_envio"].dtype
```

```python
#Media do tempo de envio por Marca
df.groupby("Marca")["Tempo_envio"].mean()
```

```python
#Verificando se existe valores nulos
df.isnull().sum()
```

Lucro por ano e por Marca

```python
#Agrupar ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum().reset_index()
lucro_ano
```

```python
#Total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Grafico total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total de produtos vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
#Grafico de lucro x ano
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum()
```

```python
#Vendas somente de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
#Lucro x mes 2009
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(title="Lucro x Mes")
plt.xlabel("Mes")
plt.ylabel("Lucro");
```

```python
#Lucro por Marca 2009
df_2009.groupby("Marca")["lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```

```python
#Lucro por classe
df_2009.groupby("Classe")["lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```
